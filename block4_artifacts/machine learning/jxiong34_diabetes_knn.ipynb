```python
#Import Libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, classification_report 
```

```python
#Import Dataset
df = pd.read_csv(r"C:\Users\jxion\OneDrive\CVTC4SEM\4-ML\Module11\diabetes.csv")

#Preview the dataset
df.info()

df.head()
```

```python
#Assign 'Outcome' as the dependent (y) variable & all other fields as independent (X) variables.
X = df.drop('Outcome', axis=1)
y = df['Outcome']
df.head()
```

```python
#Split data into test/train sets (60/40) & shuffle.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, shuffle=True, random_state=42) 
df.head()
```

```python
#Assign the KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=7)
#Train the model
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
```

```python
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
# Data point to predict
outcome = [
	6, #Pregnancies
	89, #Glucose
	66, #BloodPressure 
	23, #SkinThickness
	94, #Insulin    
	28.1, #BMI
	0.167, #DiabetesPredigreeFunction 
	21, #Age
]

# Make prediction
new_outcome = knn.predict([outcome])
new_outcome
```

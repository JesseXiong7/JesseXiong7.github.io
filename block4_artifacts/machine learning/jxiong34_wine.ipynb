```python
#Import Libraries
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
%matplotlib inline

#Import Dataset
df = pd.read_csv(r"C:\Users\jxion\OneDrive\CVTC4SEM\4-ML\Module12\winequality-red.csv")
df.head()
```

```python
#Assigning the X variables as alcohol & quality, respectively.
X = df[["alcohol", "quality"]]
X.head()
```

```python
#Fit the algorithm to the data & run predictions
model = KMeans(n_clusters = 3, init='k-means++', random_state=42)
model.fit(X)

model_predict = model.predict(X)
centroids = model.cluster_centers_
centroids
```

```python
#Plot the centroids & clusters using a scatter plot
plt.figure(figsize=(7,5))
plt.scatter(X['alcohol'], X['quality'], c=model_predict, s=50, cmap='rainbow')
plt.scatter(centroids[:, 0], centroids[:, 1], c='black', s=200, alpha=1)
plt.xlabel("Alcohol")
plt.ylabel("Quality")
plt.title("Wine Quality Clusters (k-Means)")
plt.show()
```
